# Navigating the Future of Foreign Tourism in India: Lessons Learned from the Impact of COVID-19

#### India is a country known for its rich cultural heritage, beautiful landscapes, and warm hospitality. The country has been a popular tourist destination for years, attracting visitors from all over the world. In recent years, tourism has emerged as one of the fastest-growing industries in India, contributing significantly to the country's economy. In this report, we take a comprehensive look at the different aspects of foreign tourism in India, including foreign tourist arrivals (FTAs), non-resident Indian (NRI) arrivals, international tourist arrivals (ITAs), and other related factors.

#### FTAs are a critical source of revenue for the country, and they have been steadily increasing over the past few decades. According to the available data, the number of FTAs has grown from 1.28 million in 1981 to 10.93 million in 2019. However, the COVID-19 pandemic has had a severe impact on the tourism industry globally, including India. In 2020, the number of FTAs dropped to 2.74 million, a decline of 74.9% compared to the previous year.
